```python
automatic_models = bm[bm["transmission"] == "Automatic"]# To Divide car models based on automatic transmission
manual_models = bm[bm["transmission"] == "Manual"]

print("Car models with Automatic Transmission:") # To Display the car models with automatic transmission
print(automatic_models[["model", "transmission"]])

print("\nCar models with Manual Transmission:") # To Display the car models with manual transmission
print(manual_models[["model", "transmission"]])
```

```python
Diesel_models = bm[bm["fuelType"] == "Diesel"]# To Divide car models based on fuel type
Petrol_models = bm[bm["fuelType"] == "Petrol"]
Hybrid_models = bm[bm["fuelType"] == "Hybrid"]
Other_models = bm[bm["fuelType"] == "Other"]
Electric_models = bm[bm["fuelType"] == "Electric"]

print("Car models with Diesel Models:") # To Display the car models with diesel.
print(Diesel_models[["model", "fuelType"]])

print("Car models with Petrol models:") # To Display the car models with petrol.
print(Petrol_models[["model", "fuelType"]])

print("Car models with Hybrid models:") # To Display the car models with hybrid.
print(Hybrid_models[["model", "fuelType"]])

print("Car models with other models:") # To Display the car models with others.
print(Other_models[["model", "fuelType"]])

print("Car models with Electric cars:") # To Display the car models with electric models.
print(Electric_models[["model", "fuelType"]])
```

```python
import matplotlib.pyplot as plt
```

```python
yearly_counts = bm['year'].value_counts().sort_index()
yearly_counts
```

```python
plt.figure(figsize=(10, 6)) # creating a line graph to visualize the data.
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-')
plt.title('Number of Cars by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.grid(True)
plt.show()
```

```python
fuel_type_counts = bm['fuelType'].value_counts()
fuel_type_counts
```

```python
plt.figure(figsize=(10, 6))   # creating a bar graph for number of cars w.r.t Fuel type
fuel_type_counts.plot(kind='bar')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

```python
model_fuel_counts = bm.groupby(['model', 'fuelType']).size().unstack(fill_value=0)
model_fuel_counts
```

```python
plt.figure(figsize=(12, 6))   # creating a bar graph for car models w.r.t fuel type
model_fuel_counts.plot(kind='bar', stacked=True)
plt.title('Car Models by Fuel Type')
plt.xlabel('Car Model')
plt.ylabel('Number of Cars')
plt.legend(title='Fuel Type', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
model_transmission_counts = bm.groupby(['model', 'transmission']).size().unstack(fill_value=0)
model_transmission_counts
```

```python
plt.figure(figsize=(12, 6))    # to make a bar graph of car models w.r.t transmission type.
model_transmission_counts.plot(kind='bar', stacked=True)
plt.title('Car Models by Transmission Type')
plt.xlabel('Car Model')
plt.ylabel('Number of Cars')
plt.legend(title='Transmission Type', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
# To Group the data by "Model" and calculate the average price for each car model
average_prices = bm.groupby('model')['price'].mean().sort_values(ascending=False)
average_prices
```

```python
plt.figure(figsize=(12, 6))  # to create a bar graph of different models w.r.t their average prices.
average_prices.plot(kind='bar')
plt.title('Average Price of Car Models')
plt.xlabel('Car Model')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# To Group the data by "Model" and calculate the average MPG for each car model
average_mpg = bm.groupby('model')['mpg'].mean().sort_values(ascending=False)
average_mpg
```

```python
plt.figure(figsize=(12, 6))   # to make a bar graph of different car models w.r.t their respective avearge mpg.
average_mpg.plot(kind='bar')
plt.title('Average MPG of Car Models')
plt.xlabel('Car Model')
plt.ylabel('Average MPG')
plt.xticks(rotation=45)
plt.show()
```
